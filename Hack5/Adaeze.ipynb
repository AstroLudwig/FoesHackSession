```python
from astropy.io import fits
from astropy.wcs import WCS
from astropy.coordinates import SkyCoord 
from astropy.visualization import ZScaleInterval
import matplotlib.pyplot as plt 
import astropy.units as u
```

```python
fn = 'calibrated-T21-leesha15-SN2020uea-20201011-011217-Red-BIN1-E-600-001.fit'
f = fits.open(fn)
d = f[0].data
h = f[0].header
wcs = WCS(fn)
ra = 182.858411
dec = 8.367113
Z = ZScaleInterval()
vmin,vmax = Z.get_limits(d)
x,y = wcs.all_world2pix(ra,dec,0)
x,y
```

```python
SkyCoord(ra,dec,unit=u.deg,frame='icrs').to_pixel(wcs)
```

```python
plt.imshow(d,vmin=vmin,vmax=vmax,origin='lower')
plt.scatter(x,y,c='r')
```
